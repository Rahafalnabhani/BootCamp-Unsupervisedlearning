```python
import pandas as pd
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score 
```

# Load and visualize the data

```python
#Load iris dataset :
iris = datasets.load_iris()
x = iris.data[:,2:4]        # we only take the first two features for simplicity
```

```python
# Visualize the data:
plt.scatter(x[:,0],x[:,1])
plt.xlabel('petal Length')
plt.ylabel('petal width')
plt.show()
```

# Apply K-Means Clustering

```python
# Apply kmean with k=3 :
kmeans = KMeans(n_clusters=3 , random_state=42)
kmeans.fit(x)
```

```python
# Get labels and cluster centers :
labels = kmeans.labels_
centers = kmeans.cluster_centers_
```

```python
# To call them and see :
# There are 3 classes because k=3
labels
```

```python
centers
```

# Visualize the clusters 

```python
plt.scatter(x[:,0] , x[:,1] , c=labels , cmap='viridis')
plt.scatter(centers[:,0] , centers[:,1] , c = 'red' , s=200 , alpha=0.5)
plt.xlabel('petal Length')
plt.ylabel('petal width')
plt.title('K-Mean Clustering with k=3')
plt.show()
```

# Evaluate the Model 

```python
# when the K=3 the silhouette_score is good 
score = silhouette_score(x , labels)
print('Silhouette score: ',score)
```
