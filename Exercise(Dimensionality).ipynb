```python
import pandas as pd
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the Boston Housing dataset by using the Link:
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data"
names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'PRICE']
ex = pd.read_csv(url, delim_whitespace=True, names=names)
```

```python
ex
```

```python
#Standardize the features to have mean=0 and variance=1 :
scaler = StandardScaler()
x_scaled = scaler.fit_transform(ex)
```

```python
# Apply PCA [] and fit the features selected :
pca = PCA(n_components=2)
pca.fit_transform(x_scaled)
```

```python
# Apply PCA [] and fit the features selected :
pca = PCA(n_components=2)
x_pca=pca.fit_transform(x_scaled)
```

```python
#visualizethe data
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=ex.PRICE,alpha=0.5,cmap=plt.cm.get_cmap('GnBu',10))
plt.colorbar()
plt.xlabel('first principal component')
plt.ylabel('second principal component')
plt.title('PCA on Digits Dataset')
plt.show()
```
